```python
import pandas as pd
import mysql.connector

db = mysql.connector.connect(
    host="localhost",
    user="root",
    password="secret",
    port="3306",
    database="serlo"
)

event_log = pd.read_sql("""
    select event_log.id, user.username from event_log
    join user on user.id = event_log.actor_id
    where event_log.event_id = 5 and user.username != "Legacy"
""", db)
event_log
```

```python
df = event_log.groupby("username").count()
df.rename(columns={"id": "edits"}, inplace=True)
df.sort_values(["edits"], inplace=True, ascending=False)
df = df.head(208)
df
```

```python
from IPython.display import display, Markdown, HTML

display(Markdown("# Aktuelle Profilbilder der 208 aktivsten Autor:innen"))

result = []
result.append('<div style="width: 100%; display: flex; flex-flow: row wrap; justify-content: space-around;">')

for username in df.index:
    result.append('<div style="margin: 10px">')
    result.append(f'<img src="https://community.serlo.org/avatar/{username}" style="width: 100px; height: 100px;"/>')
    result.append(f'<p>{username}</p>')
    result.append('</div>')

result.append('</div>')

display(HTML("\n".join(result)))
```
