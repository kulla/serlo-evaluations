```python
import pandas as pd
import mysql.connector

db = mysql.connector.connect(
  host="localhost",
  user="root",
  password="secret",
    port="3306",
    database="serlo"
)
pd.set_option('display.max_rows', None)

df = pd.read_sql("""
    select
        min(user.username) as username,
        min(user.id) as serlo_id,
        count(*) as edits
    from event_log
    join user on user.id = event_log.actor_id
    where
        year(event_log.date) = 2020
        and event_log.event_id = 5
    group by event_log.actor_id
    order by edits desc;
""", db)

df.to_csv("/tmp/authors2020.csv")

df
```
